```python
def delete_excess(array):
    nan_counts = np.isnan(array).sum(axis=tuple(range(array.ndim - 1)))
    if np.shape(np.shape(array))[0]==4:
        shape=np.shape(array)
        total_vox=shape[0]*shape[1]*shape[2]
        valid_indices = np.where(nan_counts != total_vox)[0]
        first=valid_indices[-1]+1
    else:
        print("Your array is the wrong fucking shape Ilana")
    return array[...,:first]
```

```python
def vox_idx_to_xyz_idxes(vox_idx: int, nx: int, ny: int, nz: int) -> tuple[int, int, int]:
    """
    Convert a linear voxel index to x, y, z indices with sanity checks.

    Parameters:
    vox_idx (int): Linear voxel index.
    nx (int): Number of voxels along the x-dimension.
    ny (int): Number of voxels along the y-dimension.
    nz (int): Number of voxels along the z-dimension.

    Returns:
    tuple[int, int, int]: Corresponding (x, y, z) indices.

    Raises:
    ValueError: If the voxel index is out of bounds.
    """
    max_vox_idx = nx * ny * nz - 1
    if not (0 <= vox_idx <= max_vox_idx):
        raise ValueError(f"Voxel index {vox_idx} is out of bounds (0 <= vox_idx <= {max_vox_idx}).")

    z = vox_idx // (nx * ny)
    rem = vox_idx % (nx * ny)
    y = rem // nx
    x = rem % nx
    return x, y, z

def xyz_idxes_to_vox_idx(x: int, y: int, z: int, nx: int, ny: int, nz: int) -> int:
    """
    Convert x, y, z indices to a linear voxel index with sanity checks.

    Parameters:
    x (int): x-coordinate index.
    y (int): y-coordinate index.
    z (int): z-coordinate index.
    nx (int): Number of voxels along the x-dimension.
    ny (int): Number of voxels along the y-dimension.
    nz (int): Number of voxels along the z-dimension.

    Returns:
    int: Linear voxel index.

    Raises:
    ValueError: If x, y, or z indices are out of bounds.
    """
    if not (0 <= x < nx):
        raise ValueError(f"x index {x} is out of bounds (0 <= x < {nx}).")
    if not (0 <= y < ny):
        raise ValueError(f"y index {y} is out of bounds (0 <= y < {ny}).")
    if not (0 <= z < nz):
        raise ValueError(f"z index {z} is out of bounds (0 <= z < {nz}).")

    return z * (nx * ny) + y * nx + x
```

```python
%time

from tqdm import tqdm

#### Loop over all voxels (io access is done nx*ny*nz times...)

for i in range(len(loom_all)):
    # subtract loom onset time for corresponding timestamps
    ts[bin_idx == i*2 + 1] -= loom_all[i]

# boolean mask of where bin_idx is odd
odd_mask = bin_idx % 2 == 1

nx, ny, nz, nt = brain.shape
n_voxels = nx * ny * nz

within_bin_brain_np = np.full((nx, ny, nz, max_len), np.nan)
within_bin_ts_rel_np = np.full((nx, ny, nz, max_len), np.nan)

for v in tqdm(range(n_voxels)):
    
    x, y, z = vox_idx_to_xyz_idxes(v, nx, ny, nz)
    
    # Collect ts_rel and brain elements that fall within loom window / bin
    within_bin_vox  = brain[x,y,z,odd_mask[x,y,z,:]]
    within_bin_vox_ts_rel = ts_rel[x,y,z,odd_mask[x,y,z,:]]

    # Get the sorted indices of the timestamp array and sort both arrays using the sorted indices
    sorted_indices = np.argsort(within_bin_vox_ts_rel)
    within_bin_vox = within_bin_vox[sorted_indices]
    within_bin_vox_ts_rel = within_bin_vox_ts_rel[sorted_indices]

    # populate the output array
    within_bin_brain_np[x,y,z,len(within_bin_brain)] = within_bin_vox
    within_bin_ts_rel_np[x,y,z,len(within_bin_brain)] = within_bin_vox_ts_rel

brain_shape = within_bin_brain_np.shape
ts_shape = within_bin_ts_rel_np.shape
print(f"Temporal filtered data shape is {brain_shape} and timestamp shape is {ts_shape}")
```

```python
def extract_roi_signal_traces(roi_ids, roi_masks, warps, hemi, signal_type):
    t0 = time.time()
    roi_time_avgs = []
    for roi in roi_ids[hemi]:
        mask = roi_masks[roi]
        masked_data = warps[:,:,:,:]*mask[np.newaxis,:,:,:] #note z-flip
        if signal_type == 'max':
            roi_time_avg = np.max(masked_data,axis=(1,2,3))
        elif signal_type == 'mean':
            roi_time_avg = np.mean(masked_data,axis=(1,2,3))
        roi_time_avgs.append(roi_time_avg)
    print(time.time()-t0)
    return np.asarray(roi_time_avgs)
```
