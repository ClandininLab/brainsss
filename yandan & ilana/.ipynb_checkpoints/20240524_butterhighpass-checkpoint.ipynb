```python
def butter_highpass(cutoff, fs, order=5):
    return butter(order, cutoff, fs=fs, btype='high', analog=False)

def butter_highpass_filter(data, cutoff, fs, order=5):
    b, a = butter_highpass(cutoff, fs, order=order)
    y = filtfilt(b, a, data)
    return y
def apply_butter_highpass(data, z, cutoff, order, fs):

    # Get the filter coefficients so we can check its frequency response.
    b, a = butter_highpass(cutoff, fs, order)
    hpf_data = butter_highpass_filter(data[:,:,z, :], cutoff, fs, order)
    return hpf_data
```

```python
def sos_butter(signal, cutoff, fs, axis=-1):
    """ Apply sos second-order butter filter to signal
    
    Args:
      signal - array like
      cutoff
      fs
      axis: filtered axis
    """
    sos = butter(2, cutoff, fs=fs, btype="highpass", analog=False, output='sos')
    return sosfiltfilt(sos, signal, axis=axis)
```
